```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("creditcard.csv")
df.head()
```

```python
df.groupby('Class').mean()
```

```python
normal_trans = df[df.Class == 0]
fraud_trans = df[df.Class == 1]
```

```python
# randomly sampling the normal transaction data 
sample_normal_trans = normal_trans.sample(n = 492) 
```

```python
# combining the sampled normal transaction and fraud transaction
new_df = pd.concat([sample_normal_trans , fraud_trans] , axis = 0)
new_df.head()
```

```python
new_df.shape
```

```python
x = new_df.drop('Class' , axis = 1)
y = new_df['Class']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=12)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
training_pred = model.predict(x_train)
print("Training Accuracy : ",accuracy_score(training_pred,y_train))
```

```python
testing_pred = model.predict(x_test)
print("Testing Accuracy : ",accuracy_score(testing_pred,y_test))
```

```python
import pickle
filename = "CCFD.sav"
pickle.dump(model,open(filename,'wb'))
```
